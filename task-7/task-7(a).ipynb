```
import cv2
import numpy as np

# Load the image
image_path = '/content/input.jpg'  # Update this to your image's path
image = cv2.imread(image_path)

# Apply median filter
median_filtered = cv2.medianBlur(image, 7)  # The second parameter is the kernel size

# Apply Gaussian blur
gaussian_blur = cv2.GaussianBlur(image, (7, 7), 0)  # The second and third parameters define the kernel size and sigma

# Save or display the results
cv2.imwrite('median_filtered.jpg', median_filtered)
cv2.imwrite('gaussian_blur.jpg', gaussian_blur)

# If you want to display the images within a Python environment (like Jupyter notebooks), you can use:
from matplotlib import pyplot as plt
plt.imshow(cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB))
plt.title('Median Filtered')
plt.show()

plt.imshow(cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB))
plt.title('Gaussian Blur')
plt.show()
```

```
import cv2

# Load the image
image_path = 'input.jpg'  # Update this to the path of your image
image = cv2.imread(image_path)
image1= cv2.imread(image_path)
# Convert the image to grayscale for better results with the function
# This step is optional and can be skipped for color images, but denoising might be more effective on grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Non-Local Means Denoising
# For a grayscale image:
denoised_gray = cv2.fastNlMeansDenoising(gray_image, None, 30, 7, 21)
# For a color image, use:
denoised_color = cv2.fastNlMeansDenoisingColored(image, None, 36, 36, 7, 30)

# Save or display the results
cv2.imwrite('denoised_gray.jpg', denoised_gray)
cv2.imwrite('denoised_color.jpg', denoised_color)

# To display the images in a Python environment like Jupyter, use:
from matplotlib import pyplot as plt
# plt.imshow(cv2.cvtColor(denoised_gray, cv2.COLOR_BGR2RGB))
# plt.title('Denoised Gray')
# plt.show()

plt.imshow(cv2.cvtColor(denoised_color, cv2.COLOR_BGR2RGB))
plt.title('Denoised Color')
plt.show()
```
